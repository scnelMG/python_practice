```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Example Plot 1 - PyPlot API Style

This is a simple ploting example

```python
y = np.array([10,20,15])
plt.plot(y)
```

```python
x = np.array([5.,10.,15.])
y = np.array([10.,20.,15.])
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'b')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'bo')
```

```python
plt.plot(x,y,'b-o',label= 'Data 1')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.legend()
```
