```python
import os
data_path = "/content/gdriv/My Drive/Colab Notebooks/data" 
data_name = os.listdir(data_path)

iris = pd.read_csv(data_path + '/' + data_name[0])
iris.head()
```

**과제1**

```python
iris['Species'].value_counts()
```

```python
idx = iris['Species'].value_counts().index
print(idx)
```

```python
ver = iris.loc[iris['Species']==idx[0],:]
vir = iris.loc[iris['Species']==idx[1],:]
seto = iris.loc[iris['Species']==idx[2],:]
```

```python
hist_ver_1 = plt.hist(ver['SepalWidthCm'], bins = 10, color = 'gray', density = False, alpha=0.4)
hist_vir_1 = plt.hist(vir['SepalWidthCm'], bins = 10, color = 'red', density = False, alpha=0.4)
hist_seto_1 = plt.hist(seto['SepalWidthCm'], bins = 10, color = 'skyblue', density = False, alpha=0.4)
labels= ["Versicolor","Virginica","Setosa"]
plt.legend(labels)
plt.title('SepalWidthCm')
```

```python
hist_ver_2 = plt.hist(ver['PetalLengthCm'], bins = 10, color = 'gray', density = False, alpha=0.5)
hist_vir_2 = plt.hist(vir['PetalLengthCm'], bins = 10, color = 'red', density = False, alpha=0.5)
hist_seto_2 = plt.hist(seto['PetalLengthCm'], bins = 10, color = 'skyblue', density = False, alpha=0.5)
labels= ["Versicolor","Virginica","Setosa"]
plt.legend(labels)
plt.title('PetalLengthCm')
```

```python
hist_ver_3 = plt.hist(ver['PetalWidthCm'], bins = 10, color = 'gray', density = False, alpha=0.5)
hist_vir_3 = plt.hist(vir['PetalWidthCm'], bins = 10, color = 'red', density = False, alpha=0.5)
hist_seto_3 = plt.hist(seto['PetalWidthCm'], bins = 10, color = 'skyblue', density = False, alpha=0.5)
labels= ["Versicolor","Virginica","Setosa"]
plt.legend(labels)
plt.title('PetalWidthCm')
```
