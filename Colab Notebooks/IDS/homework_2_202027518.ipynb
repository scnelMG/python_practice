    Hint) 출력된 값이 균등분포의 평균 및 표준 정규 분포의 평균과 유사 해야함.

**과제 1)** 특정 조건을 만족하는 Array 생성한뒤 shape print하기


```python
import numpy as np
```

```python
array_1 = np.random.rand(10)
print(array_1)
```

```python
array_1_new = array_1.reshape(10,1)
print(array_1_new)
```

```python
array_2 = np.random.randn(10)
print(array_2)
```

```python
array_2_new = array_2.reshape(10,1)
print(array_2_new)
```

```python
array_f = np.concatenate((array_1_new,array_2_new),axis=1)
```

```python
print(array_f.shape)
```

```python
print(array_f)
```

**과제 2)** 과제1을 통해 생성된 array를 열 기준으로 평균과 표준편차를 구할 것

```python
# 열 기준 평균 구하기
print(np.mean(array_f, axis=0))
```

1열은 균등분포의 평균인 0.5에 가까워야함. 

2열은 표준 정규 분포의 평균인 0에 가까워야함.

```python
# 열 기준 표준편차 구하기
print(np.std(array_f, axis=0))
```

1열은 균등분포의 표준분포인 image.png(=0.288675...)에 가까워야함.

2열은 표준 정규 분포의 표준분포인 1에 가까워야함.
