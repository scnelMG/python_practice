```python
from selenium import webdriver
import time
import csv
import requests
import bs4

# 네이버 국내증권 사이트에 접속

raw = requests.get(url)
html = bs4.BeautifulSoup(raw.text, 'html.parser')

# 업종별시세의 상승률 높은 순으로 6개
box_1 = html.find_all('table', {"class" : 'type_1'})[0]
title_1 = box_1.find_all('a', {"class" : 'tltle'})
percent_1 = box_1.find_all('span', {"class" : 'tah p11 red01'})

print("<업종별시세>\n")
for name, per in zip(title_1, percent_1):
    print(f'이름: {name.text} \n전일대비: {per.text[5:12]}')
    
# 테마별시세의 상승률 높은 순으로 6개
box_2 = html.find_all('table', {"class" : 'type_1'})[1]
title_2 = box_2.find_all('a', {"class" : 'tltle'}) 
percent_2 = box_2.find_all('span', {"class" : 'tah p11 red01'})
present = box_2.find_all('a', {"class" : 'ls'})

print("<테마별시세>\n")
for name, per, pre in zip(title_2, percent_2, present):
    print(f'이름: {name.text} \n전일대비: {per.text[5:11]} \n주도주: {pre.text} \n')
```

```python
from selenium import webdriver
import time

driver = webdriver.Chrome('./chromedriver')
url = 'https://finance.naver.com/sise/'
driver.get(url)

# 업종별시세의 상승률 높은 순으로 6개의 종목을 XPath를 이용해 가져옴
print("<<업종별시세>>\n")
for i in range(0,3):
    industry = driver.find_element_by_xpath('//*[@id="contentarea_left"]/div[1]/table/tbody/tr['+ str(i+3) + ']/td[1]/a')
    title = industry.text
    print(f"<{title}>\n")
    
    # 각 항목을 클릭하여 자세한 항목 가져오기
    industry.click()
    time.sleep(1)
    
    # 상위 주식 5개의 종목명, 현재가, 등락률 가져오기
    contents = driver.find_elements_by_css_selector('div.name_area')
    price = driver.find_elements_by_css_selector('td.number')[5::8]
    percent = driver.find_elements_by_css_selector('span.tah.p11.red01')[1:]
    for content, num, per in zip(contents[0:5], price[0:5], percent[0:5]):
        print(f"종목명: {content.text[:-2]} \n현재가: {num.text} \n등락률: {per.text}\n")
        
    #뒤로가기
    driver.back()
    
for i in range(0,3):
    industry = driver.find_element_by_xpath('//*[@id="contentarea_left"]/div[1]/table/tbody/tr['+ str(i+9) + ']/td[1]/a')
    title = industry.text
    print(f"<{title}>\n")
    
    # 각 항목을 클릭하여 자세한 항목 가져오기
    industry.click()
    time.sleep(1)
    
    # 상위 주식 5개의 종목명, 현재가, 등락률 가져오기
    contents = driver.find_elements_by_css_selector('div.name_area')
    price = driver.find_elements_by_css_selector('td.number')[5::8]
    percent = driver.find_elements_by_css_selector('span.tah.p11.red01')[1:]
    for content, num, per in zip(contents[0:5], price[0:5], percent[0:5]):
        print(f"종목명: {content.text[:-2]} \n현재가: {num.text} \n등락률: {per.text}\n")
    
    #뒤로가기
    driver.back()
    
# 테마별시세의 상승률 높은 순으로 6개의 종목을 XPath를 이용해 가져옴
print("<<테마별시세>>\n")
for i in range(0,3):
    thema = driver.find_element_by_xpath('//*[@id="contentarea_left"]/div[2]/table/tbody/tr['+ str(i+3) + ']/td[1]/a')
    title = thema.text
    print(f"<{title}>\n")
    
    # 각 항목을 클릭하여 자세한 항목 가져오기
    thema.click()
    time.sleep(1)
    
    # 상위 주식 5개의 종목명, 현재가, 등락률 가져오기
    contents = driver.find_elements_by_css_selector('div.name_area')
    price = driver.find_elements_by_css_selector('td.number')[5::8]
    percent = driver.find_elements_by_css_selector('span.tah.p11.red01')[1:]
    for content, num, per in zip(contents[0:5], price[0:5], percent[0:5]):
        print(f"종목명: {content.text[:-2]} \n현재가: {num.text} \n등락률: {per.text}\n")
    
    #뒤로가기
    driver.back()
    
for i in range(6, 8):
    thema = driver.find_element_by_xpath('//*[@id="contentarea_left"]/div[2]/table/tbody/tr['+ str(i+3) + ']/td[1]/a')
    title = thema.text
    print(f"<{title}>\n")
    
    # 각 항목을 클릭하여 자세한 항목 가져오기
    thema.click()
    time.sleep(1)
    
    # 상위 주식 5개의 종목명, 현재가, 등락률 가져오기
    contents = driver.find_elements_by_css_selector('div.name_area')
    price = driver.find_elements_by_css_selector('td.number')[5::8]
    percent = driver.find_elements_by_css_selector('span.tah.p11.red01')[1:]
    for content, num, per in zip(contents[0:5], price[0:5], percent[0:5]):
        print(f"종목명: {content.text[:-2]} \n현재가: {num.text} \n등락률: {per.text}\n")
        
    #뒤로가기
    driver.back()
a =  driver.find_element_by_xpath('//*[@id="contentarea_left"]/div[2]/table/tbody/tr[11]/td[1]/a')
title = a.text
print(f"<{title}>\n")
a.click()

contents = driver.find_elements_by_css_selector('div.name_area')
price = driver.find_elements_by_css_selector('td.number')[5::8]
percent = driver.find_elements_by_css_selector('span.tah.p11.red01')[1:]
for content, num, per in zip(contents[0:5], price[0:5], percent[0:5]):
    print(f"종목명: {content.text[:-2]} \n현재가: {num.text} \n등락률: {per.text}\n")
    
driver.close()
```
