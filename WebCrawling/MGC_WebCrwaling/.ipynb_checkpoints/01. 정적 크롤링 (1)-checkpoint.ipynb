## BeautifulSoup 사용하기

```python
import requests
import bs4

URL = 'https://dhlottery.co.kr/gameResult.do?method=byWin'
raw = requests.get(URL)
html = bs4.BeautifulSoup(raw.text, 'html.parser') # bs4.BeautifulSoup(" HTML 데이터 ", "데이터 타입" )
```

```python
target = html.find('div', {'class' : 'nums'}) # find('태그', '선택자')
# find 함수는 태그와 선택자가 일치하는 가장 첫 데이터를 리턴
print(target)
```

```python
# 1. 태그만 사용 하는 경우
html.find('div')

# 2. 선택자만 사용하는 경우
html.find(id = 'example1')
html.find(attrs = {'id':'example1'})

# 3. 태그 이름과 선택자 정보 모두 사용하는 경우
html.find('div', {'id' : 'example1'})
```

```python
balls = target.find_all("span", {'class' : 'ball_645'}) # find_all함수는 일치하는 모든 데이터를 리스트 형태로 리턴

for ball in balls:
    print(ball)
```

```python
# 깔끔하게 필요한 데이터만 출력
for ball in balls:
    print("당첨번호 : " + ball.text)
```

## 최종 결과 출력

```python
print("< 최근 로또 당첨 번호 >")
for ball in balls[:-1]:
    print("당첨번호 : ", ball.text)

print("보너스 번호 : ", balls[-1].text)
```

## Quiz (2)
BeautifulSoup 활용해서 원하는 데이터 가져오기
- 국내 증시 인기 검색 종목

```python
import requests
import bs4

URL = 'https://finance.naver.com/sise/'
raw = requests.get(URL)
html = bs4.BeautifulSoup(raw.text, 'html.parser')
```

```python
target = html.find('ul', {'class' : 'lst_pop'})
pop_stock = target.find_all("a")
```

```python
for stock, i in zip(pop_stock, range(1,11)):
        print(str(i) + '.' + stock.text)
```
