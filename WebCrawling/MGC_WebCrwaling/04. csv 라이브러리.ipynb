# csv 라이브러리
- "CSV"는 "Comma Separated Values"의 약어로, 의미 그대로 콤마로 구분된 값이라는 의미
- 즉,CSV 파일은 콤마 (',')로 구분한 텍스트 데이터 혹은 텍스트 파일

## csv 파일 작성

```python
import csv

f = open("./example.csv", 'w') # open("파일 경로" , [ 읽기/쓰기 옵션 ], [ 기타 옵션] ), 파일 작성을 위해 쓰기 옵션 'w'

wtr = csv.writer(f) # CSV 파일을 작성하는 객체 변수 'wtr' 생성

wtr.writerow(['이름', '나이', '언어']) # wtr.writerow(열에 넣을 데이터 리스트)

# 데이터 생성
name_list = ['길동', '철수', '영희']
age_list = [10, 20, 30]
lan_list = ['파이썬', 'C', '자바']

# 각 행에 데이터 작성
for i in range(3):
    name = name_list[i]
    age = age_list[i]
    language = lan_list[i]
    wtr.writerow([name, age, language])
    
# 파일 닫기
f.close()
```

```python
import csv

# 옵션 <newline = ''> 추가, 없으면 한 행씩 건너뛰면서 데이터가 작성되는 문제가 발생
f = open('./example2.csv', 'w', newline = '')

wtr = csv.writer(f)

wtr.writerow(['이름', '나이', '언어'])

all_name = ['길동', '철수', '영희']
all_age = [10, 20, 30]
all_language = ['파이썬', 'C', '자바']

for i in range(3):
    name = all_name[i]
    age = all_age[i]
    language = all_language[i]
    wtr.writerow([name, age, language])

f.close()
```

## csv 파일 읽기

```python
import csv

# 읽어올 'example.csv' 파일을 변수 'f'에 저장
f = open('./example2.csv', 'r') # 파일 속 데이터를 읽어오기 때문에, 읽기 옵션을 'r'로 설정

# CSV 파일의 모든 데이터를 변수 'rdr'에 저장
rdr = csv.reader(f)

# next(rdr) # 열의 제목은 필요하지 않는다면 next 함수 사용
# 'rdr'에 저장된 데이터를 행별로 출력
for row in rdr:
    print(row)

f.close()
```

## 저장된 csv 파일에 행 추가

```python
import csv

# 추가 모드 옵션 'a' 사용
f = open('./example2.csv', 'a', newline = '') # 수정을 위해 추가 모드 옵션 'a' 사용

wtr = csv.writer(f)

wtr.writerow(['바둑', 40, '파이썬'])
wtr.writerow(['오목', 50, 'C'])

f.close()
```
