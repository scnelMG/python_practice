# n번째 분수

## 문제


아래 그림과 같이, 배열에 다음과 같은 분수가 적혀있습니다. 

그림처럼 지그재그 형태로 순서를 정할 경우, 1/1은 첫 번째, 1/2는 두 번째 분수가 됩니다. 

n번째 분수를 구하는 프로그램을 작성해보세요.



```python
goal = int(input("원하는 n 값을 입력해주세요:"))

line = 1 # 몇 번째 line에 위치하느냐를 구하기
num = goal

while num > line:
    num -= line # while문을 빠져나가면 특정 라인의 몇번쨰에 위치하는지 알 수 있음.
    line += 1

print(f"{line}반째 라인에 원하는 {goal}번쨰 분수가 존재합니다.")

if line % 2 == 0: # 라인이 짝수 번쨰 라인일 때
    a = num 
    b = line - num + 1
else: # 라인이 홀수 번쨰 라인일 때
    a = line - num + 1
    b = num
    
print(f"{goal}번째 분수의 값은 {a}/{b} 입니다.")
```
