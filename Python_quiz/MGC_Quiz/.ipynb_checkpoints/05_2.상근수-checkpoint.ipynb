# 상근수

## 문제


양의 정수 n의 각 자리수 제곱의 합을 계산한다. 그렇게 해서 나온 합도 각 자리수의 제곱의 합을 계산한다.

이렇게 반복해서 1이 나온다면, 해당 수는 상근수 라고 한다.



    


    


위 식에서 700은 상근수이고 2는 상근수가 아니다.


양의 정수 N이 입력으로 주어졌을 때, N보다 작거나 같은 모든 상근수를 구해 출력 해 보세요.

```python
def find(num):
    num = str(num)
    number=[]
    number.append(num)
    for n in number:  
        result=0
        for i in range(len(n)):
            a = n[i]
            result=result + (int(a) ** 2)
        result=str(result)
        if result == '1':
            return True
        elif result in number:
            break
        number.append(result)
```

```python
find('700')
```

```python
num = int(input("정수를 입력하세요: "))
```

```python
for i in range(int(num)):
    if find(i) == True:
        print(i, end =' ')
```
