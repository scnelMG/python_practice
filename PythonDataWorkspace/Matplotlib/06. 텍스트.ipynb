# 6. 텍스트

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # MAc
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, marker = 'o')

for idx, txt in enumerate(y): # idx = 인덱스, txt in enumerate(y) = y의 값을 txt에 넣음
    plt.text(x[idx], y[idx] + 0.3, txt, ha = 'center', color = 'blue') # txt가 들어갈 위치 설정, ha = 가운데 정렬
```
