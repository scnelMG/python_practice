# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # MAc
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth = 5) # 선 굵기 설정
```

## 마커 (marker)

```python
plt.plot(x, y, marker = 'o') # 데이터 값 표시
```

```python
plt.plot(x, y, marker = 'o', linestyle = 'None') # 점만 표시
```

```python
plt.plot(x, y, marker = 'v' ) # 삼각형으로 표시
```

```python
plt.plot(x, y, marker = 'v' , markersize = 10) # 삼각형으로 표시, 삼각형 크기 설정
```

```python
plt.plot(x, y, marker = 'X' , markersize = 10) # X로 표시, X 크기 설정
```

```python
plt.plot(x, y, marker = 'X' , markersize = 10, markeredgecolor = 'red') # 마커의 겉부분 색 변경
```

```python
plt.plot(x, y, marker = 'X' , markersize = 10, markerfacecolor = 'yellow') # 마커의 안쪽 부분 색 변경
```

## 선 스타일

```python
plt.plot(x, y, linestyle = ':') # 선 모양을 점선으로
```

```python
plt.plot(x, y, linestyle = '--') # 선 모양을 더 두꺼운 점선으로
```

```python
plt.plot(x, y, linestyle = '-.') # 선 모양을 선, 점을 반복한 것으로
```

```python
plt.plot(x, y, linestyle = '-') # 선 모양을 일반 실선으로
```

## 색깔

```python
plt.plot(x, y, color = 'pink')
```

```python
plt.plot(x, y, color = '#ff0000')
```

```python
plt.plot(x, y, color = 'b')
```

```python
plt.plot(x, y, color = 'g')
```

## 포맷

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go')
```

## 축약어

```python
plt.plot(x, y, marker = 'o', mfc = 'red', ms=10, mec = 'blue', ls=':' )
# mfc = markerfacecolor,  ms = markersize, mec =  markeredgecolor, ls = linestyle
```

## 투명도

```python
plt.plot(x, y, marker = 'o', mfc = 'red', ms=10, mec = 'blue', ls=':' , alpha = 0.3) # alpha : 투명도(0~1)
```

## 그래프 크기

```python
plt.figure(figsize = (10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize = (10,5), dpi = 100) # dots per inche --확대
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor = 'yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor = '#a1c3ff')
plt.plot(x, y)
```
