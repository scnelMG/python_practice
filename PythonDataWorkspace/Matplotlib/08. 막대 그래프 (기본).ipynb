```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # MAc
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt. bar(labels, values, color='r')
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

colors =['r', 'g', 'b']
plt. bar(labels, values, color = colors, alpha = 0.7)
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt. bar(labels, values)
plt.ylim(175, 195) # 원하는 y값 범위 설정
```

```python
plt. bar(labels, values, width = 0.5) # 그래프 두꺠 설정
```

```python
plt. bar(labels, values, width = 0.3)
plt.xticks(rotation = 45) # x축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation = 45) # y축의 이름 데이터 각도를 45도로 설정
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks)
```
